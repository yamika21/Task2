Data Preprocessing

1. Data Preprocessing and Exploratory Data Analysis (EDA):

```python
import pandas as pd
df = pd.read_csv('heart.csv')
```

Handle Missing Values: Identify and handle missing data using pandas.

```python
df.isnull().sum()  # Check for missing values
df.fillna(df.mean(), inplace=True)  # Replace missing values with mean (or use another strategy)
```

Normalize Data: If needed, normalize numerical features using MinMaxScaler or StandardScaler from sklearn.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
```

Convert Categorical Features: If your dataset contains categorical data, use one-hot encoding or label encoding.

```python
df = pd.get_dummies(df, drop_first=True)  # One-hot encoding
```

Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

Visualizing Distributions

```python
# Distribution of 'age' feature
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Distribution of 'chol' (cholesterol level)
sns.histplot(df['chol'], kde=True)
plt.title('Cholesterol Level Distribution')
plt.show()
```

Target Variable Distribution

```python
# Check how the target variable is distributed
sns.countplot(x='target', data=df)
plt.title('Heart Disease vs No Heart Disease')
plt.show()
```
