Data Preprocessing and Exploratory Data Analysis (EDA):

Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

Step 2: Load the Dataset

```python
# Load the dataset
df = pd.read_csv('heart.csv')  # Make sure to adjust the file path if necessary
```

Step 3: Explore the Dataset

```python
# Display the first few rows of the dataset
df.head()

# Display general info about the dataset (data types, missing values, etc.)
df.info()

# Summary statistics of the dataset
df.describe()
```

Step 4: Data Preprocessing
Handling Missing Values

```python
# Check for missing values
df.isnull().sum()

# Fill missing values with the median (or you can choose other strategies)
df.fillna(df.median(), inplace=True)
```

Encoding Categorical Variables

```python
# If there are categorical columns, convert them into numeric using One-Hot Encoding
df = pd.get_dummies(df, drop_first=True)
```

Normalize the Data (Optional)

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop('target', axis=1))

# Convert scaled data back to DataFrame
df_scaled = pd.DataFrame(scaled_data, columns=df.drop('target', axis=1).columns)
df_scaled['target'] = df['target']
```

Step 5: Exploratory Data Analysis (EDA)
Visualizing the Correlation Heatmap

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
